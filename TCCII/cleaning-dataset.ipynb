```python
import csv
import pandas as pd
```

```python
reddit_questions = pd.read_csv('data/reddit/reddit_questions.csv', on_bad_lines='skip',  sep = ';')
reddit_answers = pd.read_csv('data/reddit/reddit_answers.csv', on_bad_lines='skip',  sep = ';')
```

```python
reddit_questions.head()
```

```python
reddit_answers.head()
```

```python
def insertRedditData(writer):
    for index, row in reddit_questions.iterrows():
        #Get the answers for the question
        r_answers = reddit_answers.loc[reddit_answers['q_id'] == row['id']]

        #Remove the invalid answers
        r_answers = r_answers.drop(r_answers[r_answers.text.str.len() < 15].index)

        #Get the best voted
        r_answers = r_answers[r_answers['votes'] > r_answers.loc[:, 'votes'].mean()]

        for i, r in r_answers.iterrows(): 
            writer.writerow([row['text'], r['text']])
        
def insertInteractionsData(writer):
    filename = 'data/interactions.csv'

    with open(filename, 'r') as interactionsfile:
        datareader = csv.reader(interactionsfile)
        for row in datareader:
            writer.writerow([row[0], row[1]])
        
```

```python
csvfile = open('data/dataset.csv', 'w')
writer = csv.writer(csvfile)
writer.writerow(['question', 'answer'])

insertInteractionsData(writer)
insertRedditData(writer)

csvfile.close()
```
